```python
import pandas as pd
```

```python
from sklearn.datasets import load_wine
data = load_wine()
df_X = pd.DataFrame(data.data, columns=data.feature_names) # dataframeに変換
df_X.head() # 最初の数行を出力
```

```python
df_y = pd.DataFrame(data.target, columns=["kind(target)"]) # 目的変数をdataframeに変換
```

```python
df_y.head()
```

```python
# データを連結させて見やすくする
df = pd.concat([df_X, df_y], axis=1)
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.hist(df.loc[:, "alcohol"]) # locはlocationのことでcolumn_nameで指定可能
```

```python
plt.boxplot(df.loc[:, "alcohol"])
```

```python
# corrメソッドを使い相関係数をまとめて計算し出力
df.corr()
```

```python
# describeメソッドで列ごとに統計情報を出力
df.describe()
```

```python
# scatter_matrixを用いて散布図行列を出力
from pandas.plotting import scatter_matrix
_ = scatter_matrix(df, figsize=(15, 15))
```

```python
# 列を限定して出力
_ = scatter_matrix(df.iloc[:, [0, 9, -1]])
```
