```python
%matplotlib inline # showメソッドを後で実行しなくてもグラフが出力されるおまじない
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(-5, 5, 101) # -5から5まで101個のデータを生成
y1 = np.sin(x1) # sin関数を用いてデータを生成
```

```python
plt.plot(x1, y1) # 簡易出力
```

```python
fig, ax = plt.subplots()
ax.set_title("Sin")
ax.set_xlabel("rad")
ax.plot(x1, y1) # グラフ出力
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
plt.show() # jupyter notebook以外でもグラフが出力できるように記載
```

```python
x2 = np.arange(100) # 0-99の配列を代入
y2 = x2 * np.random.rand(100)
```

```python
# 散布図
plt.scatter(x2, y2)
```

```python
# ヒストグラム
plt.hist(y2, bins=5)
```

```python
# 棒グラフ
plt.bar(x2, y2)
```

```python
# 折れ線グラフ
plt.plot(x2, y2)
```

```python
# 箱ひげ図
plt.boxplot(y2)
```

```python
# scikit-learnに内包されているワインデータを可視化
from sklearn.datasets import load_wine
data = load_wine() # ワインデータをロードしてdataに格納

x3 = data.data[:, [0]] # indexに対応するカラムの抽出
y3 = data.data[:, [9]]

plt.scatter(x3, y3)
```
